```python
import numpy as np
import matplotlib.pyplot as plt

for variant in ['bh1', 'bh2']:
    for NFE in [5, 10, 15, 20, 25, 30, 35, 40]:
        for corrector_order in [2, 3]:
            file = f"/data/guided-diffusion/scale/unipc_rbf/variant={variant},NFE={NFE},predictor_order=2,corrector_order={corrector_order}.npz"
            if os.path.exists(file):
                data = np.load(file)
                plt.figure(figsize=[5, 3])
                plt.title(f"{variant}, {NFE}, {corrector_order}")
                plt.plot(data['optimal_log_scales'][:, 1:].T)
                plt.show()
```
