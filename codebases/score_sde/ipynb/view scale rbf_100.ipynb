```python
import numpy as np
NFE = 5
npz_file = f'/data/score_sde_scale_rbf_100/NFE={NFE},p=100.npz'
#npz_file = f'/data/score_sde_scale_rbf/NFE={NFE},p=3.npz'
data = np.load(npz_file)
print(list(data.keys()))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=[5, 3])
plt.title(f"NFE={NFE}")
plt.plot(data['optimal_log_scales'].T)
plt.grid()
plt.show()
```

```python
log_s = np.linspace(-2.0, 2.0, 100)
rows = len(data['pred_losses_list'])//5
plt.figure(figsize=[18, rows*5])
for i, (pred_losses, corr_losses) in enumerate(zip(data['pred_losses_list'], data['corr_losses_list'])):
    plt.subplot(rows+1, 5, i+1)
    plt.title(f"step: {i+1}")
    plt.plot(log_s, pred_losses)
    plt.plot(log_s, corr_losses)
    
    plt.grid()
    plt.ylim([0, 0.2])
plt.show()
```
