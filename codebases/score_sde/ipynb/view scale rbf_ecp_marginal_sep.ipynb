```python
import os
import numpy as np
import matplotlib.pyplot as plt

NFEs = [5, 10, 15, 20, 25, 30, 35, 40]
root_dirs = ['/data/score_sde/scale/rbf_ecp_marginal',
            #  '/data/score_sde/scale/rbf_ecp_marginal4',
            #  '/data/score_sde/scale/rbf_ecp_marginal5',
            #  '/data/score_sde/scale/rbf_ecp_marginal6',
             '/data/score_sde/scale/rbf_ecp_marginal_sep',
             ]

for root_dir in root_dirs:
    print(root_dir)
    plt.figure(figsize=[18, 3])
    for i, NFE in enumerate(NFEs):
        if 'marginal4' in root_dir:
            npz_file = os.path.join(root_dir, f'NFE={NFE},p=4.npz')
        elif 'marginal5' in root_dir:
            npz_file = os.path.join(root_dir, f'NFE={NFE},p=5.npz')
        elif 'marginal6' in root_dir:
            npz_file = os.path.join(root_dir, f'NFE={NFE},p=6.npz')        
        else:
            npz_file = os.path.join(root_dir, f'NFE={NFE},p=3.npz')
        data = np.load(npz_file)
        plt.subplot(1, len(NFEs), i+1)
        plt.title(f"NFE={NFE}")
        plt.plot(data['optimal_log_scales'][0].T, label='predictor')
        plt.plot(data['optimal_log_scales'][1].T, label='corrector')
        plt.grid()
        plt.legend()
    plt.show()
        
```
