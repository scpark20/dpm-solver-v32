```python
import numpy as np
NFE=25
npz_file = f'/data/score_sde_scale_rbf_mix/NFE={NFE},p=3.npz'
data = np.load(npz_file)
print(list(data.keys()))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=[5, 3])
plt.title(f"NFE={NFE}")
plt.plot(data['optimal_log_scales'][0].T, label='noise p')
plt.plot(data['optimal_log_scales'][2].T, label='noise c')
plt.grid()
plt.legend()
plt.show()

plt.figure(figsize=[5, 3])
plt.title(f"NFE={NFE}")
plt.plot(data['optimal_log_scales'][1].T, label='data p')
plt.plot(data['optimal_log_scales'][3].T, label='data c')
plt.grid()
plt.legend()
plt.show()
```

```python
log_s = np.linspace(-2.0, 2.0, 100)
rows = len(data['pred_losses_noise_list'])//5
plt.figure(figsize=[18, rows*5])
for i, (noise_losses, data_losses) in enumerate(zip(data['pred_losses_noise_list'], data['corr_losses_noise_list'])):

    plt.subplot(rows+1, 5, i+1)
    plt.title(f"step: {i+1}")
    plt.plot(log_s, data_losses)
    plt.plot(log_s, noise_losses)
    plt.grid()
    #plt.ylim([0, 0.5])
plt.show()
```

```python
log_s = np.linspace(-2.0, 2.0, 100)
rows = len(data['pred_losses_data_list'])//5
plt.figure(figsize=[18, rows*5])
for i, (noise_losses, data_losses) in enumerate(zip(data['pred_losses_data_list'], data['corr_losses_data_list'])):

    plt.subplot(rows+1, 5, i+1)
    plt.title(f"step: {i+1}")
    plt.plot(log_s, data_losses)
    plt.plot(log_s, noise_losses)
    plt.grid()
    #plt.ylim([0, 0.5])
plt.show()
```
