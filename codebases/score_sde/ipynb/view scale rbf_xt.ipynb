```python
import numpy as np
NFE=30
npz_file = f'/data/score_sde_scale_rbf_xt/NFE={NFE},p=3.npz'
data = np.load(npz_file)
print(list(data.keys()))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=[5, 3])
plt.title(f"NFE={NFE}")
plt.plot(data['optimal_log_scales'].T)
plt.grid()
plt.show()
```

```python
log_s = np.linspace(-2.0, 2.0, 100)

for i, pred_losses in enumerate(data['pred_losses_list']):
    plt.figure(figsize=[5, 3])
    plt.title(f"step: {i}")
    plt.plot(log_s, pred_losses)
    plt.grid()
    plt.show()
```

```python
log_s = np.linspace(-2.0, 2.0, 100)
plt.figure(figsize=[5, 3])
for i, pred_losses in enumerate(data['corr_losses_list']):
    #plt.title(f"step: {i}")
    plt.plot(log_s, pred_losses)
plt.grid()
plt.show()
```
