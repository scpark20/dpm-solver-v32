```python
import numpy as np
npz_file = ''
data = np.load('/data/score_sde_outputs/checkpoint_8/euler_100/samples_0.npz')
print(list(data.keys()))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 이미 로드된 data 딕셔너리를 가정
samples = data['samples_raw']  # (256, 3, 32, 32)
samples = samples.clip(0, 1)

# 10x10 그리드 생성
fig, axes = plt.subplots(nrows=10, ncols=10, figsize=(10, 10))

# 앞의 100장을 10x10 격자로 표시
for idx, ax in enumerate(axes.flatten()):
    # idx가 0부터 99까지 반복
    img = samples[idx]         # (3, 32, 32)
    img = np.transpose(img, (1, 2, 0))  # (32, 32, 3)으로 변환
    
    # 값 범위가 -1~1인 경우 0~1 범위로 스케일링 예시 (필요 없으면 생략 가능)
    # img = (img - img.min()) / (img.max() - img.min())
    
    ax.imshow(img)
    ax.axis('off')

plt.tight_layout()
plt.show()
```
