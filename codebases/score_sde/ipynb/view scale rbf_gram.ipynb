```python
import numpy as np
NFE=20
npz_file = f'/data/score_sde_scale_rbf_gram/NFE={NFE},p=3.npz'
#npz_file = f'/data/score_sde_scale_rbf/NFE={NFE},p=3.npz'
data = np.load(npz_file)
print(list(data.keys()))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=[5, 3])
plt.title(f"NFE={NFE}")
plt.plot(data['optimal_log_scales'].T)
plt.grid()
plt.show()
```

```python
log_s = np.linspace(-2.0, 2.0, 100)
rows = len(data['pred_losses_list'])//5
plt.figure(figsize=[18, rows*5])
for i, (data_losses, noise_losses, losses) in enumerate(zip(data['pred_data_losses_list'], data['pred_noise_losses_list'], data['pred_losses_list'])):
    r = 0.999
    data_losses = data_losses * r
    noise_losses = noise_losses * (1-r)
    losses = data_losses + noise_losses
    plt.subplot(rows+1, 5, i+1)
    plt.title(f"step: {i+1}")
    #plt.plot(log_s, data_losses)
    #plt.plot(log_s, noise_losses)
    plt.plot(log_s, losses)
    plt.grid()
    #plt.ylim([0, 0.5])
plt.show()
```
