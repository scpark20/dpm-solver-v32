```python
import numpy as np
import matplotlib.pyplot as plt

log_s = np.linspace(0.0, 3.0, 100)
plt.figure(figsize=[5, 3])

for NFE in [5, 6, 8, 10, 12, 15, 20, 25]:
    npz_file = f'/data/score_sde_scale_rbf_const/NFE={NFE},p=3.npz'
    data = np.load(npz_file)

    # plt.plot    line  
    line, = plt.plot(log_s, data['losses'], label=f'NFE={NFE}')
    
    # optimal_log_scale    (   )
    plt.axvline(data['optimal_log_scale'], color=line.get_color(), linestyle='--')
    print(data['optimal_log_scale'])

plt.grid()
plt.ylim([0, 0.02])
plt.legend()
plt.show()
```
