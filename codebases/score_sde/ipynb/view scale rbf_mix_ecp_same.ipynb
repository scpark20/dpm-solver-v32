```python
import numpy as np
NFE=25
npz_file = f'/data/score_sde_scale_rbf_mix_ecp_same/NFE={NFE},p=3.npz'
data = np.load(npz_file)
print(list(data.keys()))

NOISE_PRED = 0
NOISE_CORR = 1
DATA_PRED = 2
DATA_CORR = 3
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=[5, 3])
plt.title(f"NFE={NFE}")
plt.plot(data['optimal_log_scales'][NOISE_PRED].T, label='noise p')
plt.plot(data['optimal_log_scales'][NOISE_CORR].T, label='noise c')
plt.grid()
plt.legend()
plt.show()

plt.figure(figsize=[5, 3])
plt.title(f"NFE={NFE}")
plt.plot(data['optimal_log_scales'][DATA_PRED].T, label='data p')
plt.plot(data['optimal_log_scales'][DATA_CORR].T, label='data c')
plt.grid()
plt.legend()
plt.show()
```

```python
log_s = np.linspace(-2.0, 2.0, 33)
rows = len(data['noise_loss_line_list'])//5
plt.figure(figsize=[18, rows*5])
for i, (noise_losses, data_losses) in enumerate(zip(data['noise_loss_line_list'], data['data_loss_line_list'])):

    plt.subplot(rows+1, 5, i+1)
    plt.title(f"step: {i+1}")
    plt.plot(log_s, data_losses)
    plt.plot(log_s, noise_losses)
    plt.grid()
    #plt.ylim([0, 0.5])
plt.show()
```
