```python
import numpy as np
import matplotlib.pyplot as plt

NFEs = [5, 6, 8, 10, 12, 15]
for scale in [1.5, 3.5, 5.5, 7.5]:
    plt.figure(figsize=[18, 3])
    for i, NFE in enumerate(NFEs):
        npz_file = f'/data/ldm/dc{scale}/NFE={NFE},p=3,number=0.npz'
        data = np.load(npz_file)
        plt.subplot(1, len(NFEs), i+1)
        plt.title(f"{scale}, {NFE}")
        plt.plot(data['dc_ratios'])
        plt.ylim(0, 2)
    plt.show()
```
